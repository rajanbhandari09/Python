This Ipython notebook contains code samples for:
- Mathematical operations like cosine, sine on numpy arrays
- Plotting numpy array values on 2D axes

```python
import numpy as np
import matplotlib.pyplot as pp
```

```python
%matplotlib inline
```

```python
x = np.linspace(0,10,40)
x
```

```python
cosx = np.cos(x)
cosx
```

```python
pp.plot(x,cosx)
```

```python
sinx = np.sin(x)
sinx
```

```python
pp.plot(x,sinx)
```

```python
pp.plot(x,sinx)
pp.plot(x,cosx)
```

```python
pp.plot(x,sinx,'x')
pp.plot(x,cosx,'o')
```

```python
y = sinx*cosx
z = cosx**2 - sinx**2
pp.plot(x,y)
pp.plot(x,z)
pp.legend(['sinxcosx','cosx^2-sinx^2'])
```
